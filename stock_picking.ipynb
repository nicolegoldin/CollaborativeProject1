```python
import pandas as pd
from pathlib import Path
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stock_csv_path = Path('monthly_data.csv')
monthly_df = pd.read_csv(stock_csv_path, index_col = 'Date')
returns = monthly_df.pct_change()
returns.dropna(inplace = True)
```

```python
correlations = monthly_df.corr()
sns.heatmap(correlations, cmap = 'Reds', vmax = 1, vmin = 0).set_title("Correlation Matrix")
```

```python
std_dev = returns.std()
monthly_std_dev = std_dev*(12**0.5)
```
