```python
import pandas as pd
from pathlib import Path
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stock_csv_path = Path('monthly_data.csv')
monthly_df = pd.read_csv(stock_csv_path, index_col = 'Date')
monthly_df.head(3)
```

```python
returns = monthly_df.pct_change()
returns.dropna(inplace = True)
returns.head(3)
```

```python
correlations = monthly_df.corr()
correlations.head(3)
```

```python
sns.heatmap(correlations, cmap = 'Blues', vmax = 1, vmin = 0).set_title("Correlation Matrix")
```
